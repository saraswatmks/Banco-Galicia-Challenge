```python
# Drop the duplicate columns
pageviews = pageviews.drop(['CONTENT_CATEGORY_BOTTOM'], axis=1)
```

```python
X_test = []
for c in pageviews.drop(["USER_ID", "FEC_EVENT"], axis=1).columns:
    print("haciendo", c)
    temp = pd.crosstab(pageviews.USER_ID, pageviews[c])
    temp.columns = [c + "_" + str(v) for v in temp.columns]
    X_test.append(temp)
    # X_test.append(temp.apply(lambda x: x / x.sum(), axis=1))
    # X_test.append(temp)
X_test = pd.concat(X_test, axis=1)
```

```python
X_test.head()
```

```python
pageviews = pageviews[pageviews.FEC_EVENT.dt.month < 10]
X_train = []
for c in pageviews.drop(["USER_ID", "FEC_EVENT"], axis=1).columns:
    print("haciendo", c)
    temp = pd.crosstab(pageviews.USER_ID, pageviews[c])
    temp.columns = [c + "_" + str(v) for v in temp.columns]
    X_train.append(temp)
    #X_train.append(temp.apply(lambda x: x / x.sum(), axis=1))
X_train = pd.concat(X_train, axis=1)
```

```python
features = list(set(X_train.columns).intersection(set(X_test.columns)))
X_train = X_train[features]
X_test = X_test[features]
```

```python
device_data_train = device_data[device_data['FEC_EVENT'].dt.month < 10]

device_data_train = pd.crosstab(device_data_train['USER_ID'],
                           [device_data_train['IS_MOBILE_DEVICE'], device_data_train['CONNECTION_SPEED']])

cols = ['device' + '_' + str(x) + '_' + str(y) 
        for x, y in zip(device_data_train.columns.get_level_values(0), device_data_train.columns.get_level_values(1))]
device_data_train.columns = cols

device_data_train = device_data_train.apply(lambda x: x / x.sum(), axis=1)
device_data_train.head()
```

```python
device_data = pd.crosstab(device_data['USER_ID'],
                           [device_data['IS_MOBILE_DEVICE'], device_data['CONNECTION_SPEED']])

cols = ['device' + '_' + str(x) + '_' + str(y) 
        for x, y in zip(device_data.columns.get_level_values(0), device_data.columns.get_level_values(1))]
device_data.columns = cols

device_data = device_data.apply(lambda x: x / x.sum(), axis=1)
device_data.head()
```

```python
device_data_train.to_csv('data/device_train.csv', index=True, compression='gzip')
device_data.to_csv('data/device_test.csv', index=True, compression='gzip')
```
